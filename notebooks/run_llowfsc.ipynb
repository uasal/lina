```python
def take_im_fun(
        CAMLO_STREAM,
        NFRAMES=1,
    ):
    camlo_im = np.mean(CAMLO_STREAM.grab_many(NFRAMES), axis=0) if NFRAMES>1 else CAMLO_STREAM.grab_after(1,0)[0]
    return camlo_im

take_im_params = {
    'CAMLO_STREAM':CAMLO_STREAM,
    'NFRAMES':1,
}

def set_dm_fun(
        command,
        DM_STREAM,
    ):

    DM_STREAM.write(ensure_np_array(command))

set_dm_params = {
    'DM_STREAM':ImageStream('dmdisp01sim'),
}

def get_dm_fun(
        DM_STREAM,
    ):

    return DM_STREAM.grab_latest()


get_dm_params = {
    'DM_STREAM':ImageStream('dmdisp01sim'),
}
```

```python
def compute_zpo(
        DM_STREAMS,
        dm_mask,
        wfs_mask,
        response_matrix,
        dm_modal_matrix,
        ZPO_STREAM,
    ):
    zpo = np.zeros((wfs_mask.shape[0], wfs_mask.shape[1]))
    for i in range(len(DM_STREAMS)):
        zpo[wfs_mask] += response_matrix.dot( dm_modal_matrix.dot(DM_STREAMS[i].grab_latest()[dm_mask])/952 )

    ZPO_STREAM.write(zpo)
    return zpo

def get_ffo(
        DM_STREAMS,
        dm_mask,
        dm_modal_matrix,
    ):
    ffo = 0.0
    for i in range(len(DM_STREAMS)):
        ffo += dm_modal_matrix.dot(DM_STREAMS.grab_latest()[dm_mask])
    return ffo
```

```python
zpo_params = {
    'DM_STREAMS':DM_STREAMS[2:],
    'dm_mask':dm_mask,
    'wfs_mask':wfs_mask,
    'response_matrix':zer_rm,
    'dm_modal_matrix':dm_zernike_modes[:, dm_mask],
    'ZPO_STREAM':ZPO_STREAM,
}

compute_zpo(**zpo_params)

zpo = ZPO_STREAM.grab_latest()

utils.imshow(
    [zpo]
)
```

```python
reload(llowfsc)

gains = np.array(2*[0.8] + 8*[0.05])

take_im_params.update({
    'NFRAMES':1,
})

times = []
try:
    print('Running LLOWFSC simulator continuously ...')
    while 1:
        llowfsc.run(
            take_im_fun,
            take_im_params,
            set_dm_fun,
            set_dm_params,
            get_dm_fun,
            get_dm_params,
            ref_im,
            P,
            dm_zernike_modes,
            wfs_mask,
            gains,
            dark_im=camlo_dark,
            get_zpo=compute_zpo,
            get_zpo_params=zpo_params,
        )
        times.append(time.time())
except KeyboardInterrupt:
    set_dm_fun(np.zeros((Nact,Nact)), **set_dm_params)
    print('LLOWFSC simulator terminated.')
```
